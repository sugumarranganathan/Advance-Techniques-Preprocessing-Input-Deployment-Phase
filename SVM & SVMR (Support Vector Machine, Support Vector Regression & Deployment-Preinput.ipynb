Your output
array([33], dtype=int32)
Meaning:

**Your model is using 33 support vectors**

----------

What is a support vector?

In SVM / SVR:

model draws a best boundary / best fitting margin
only some important data points decide that boundary
those important points are called support vectors

 These are the key training points that influence the model most.

SVR (Support Vector Regression) is a machine learning algorithm used to predict continuous values like salary, price, marks, temperature, etc.

Simple meaning:
SVM = used for classification (Yes/No, Cat/Dog)
SVR = used for regression (predicting a number)
Example:

If you give:

experience = 5 years
skill = 8

SVR can predict:

salary = ₹50,000
One-line:

SVR is used when the output is a number, not a category.

**If output is class/category → use SVM**
If output is continuous number → use SVR
